```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8]) # control the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
#draw some subgraphs
fig,axes=plt.subplots(nrows=1,ncols=2) #submap is of 1 row, 2 cols
for ax in axes:
    ax.plot(x,y,'r')        
```

```python
# we can also draw a picture or graph inside another graph
fig= plt.figure()

# control the left, right, width, height of the canvas (from 0 to 1)

axes1= fig.add_axes([0.1,0.1,0.8,0.8]) #big canvas
axes2= fig.add_axes([0.2,0.5,0.3,0.3]) #small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig,ax=plt.subplots(dpi=100)
#line width
ax.plot(x,x+1,color='blue',lw=2,linestyle='-')
ax.plot(x,x+2,color='red',lw=2,linestyle=':')
ax.plot(x,x+3,color='green',lw=2,linestyle='dotted')

line, =ax.plot(x,x+4,color='black',lw=1.5,linestyle='-.')
line.set_dashes([10,20,20,10])
```

```python
fig,ax=plt.subplots(dpi=100)
#line width
ax.plot(x,x+1,color='blue',lw=2,marker='o')
ax.plot(x,x+2,color='red',lw=2,marker='s')
ax.plot(x,x+3,color='green',lw=2,marker='+')
ax.plot(x,x+4,color='black',lw=1.5,marker='1')

ax.plot(x,x+5,color='blue',lw=2,marker='o',markersize=3)
ax.plot(x,x+6,color='red',lw=2,marker='s',markerfacecolor='black')
```

```python
# Set the canvas grid and axis range
fig,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)

axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
axes[1].grid(True)
```

```python
# Other 2D Graphs
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].set_title('scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n,n**2,lw=2,color='red')

axes[2].set_title('bar')
axes[2].bar(n,n**2,align='center',alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(x,x**2,x**3,color='green')
```

```python
# Draw a radar chart
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,0.6,0.6],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,lw=3)
```

```python
#Draw an Histogram
n=np.random.randn(10000)
fig,axes=plt.subplots(1,2,figsize=(12,4))

axes[0].set_title('Default histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram')
axes[1].hist(n,cumulative=True,bins=8)
```

```python
#Draw contour image
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

#Define x,y,z
delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
x,y=np.meshgrid(x,y)
z1=np.exp(-x**2-y**2)
z2=np.exp(-(x-1)**2-(y-1)**2)
z=(z1-z2)*2

#Draw contour with x,y,z values
fig,ax=plt.subplots()
CS=ax.contour(x,y,z)
ax.clabel(CS,fontsize=10)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig=plt.figure(figsize=(14,6))
#specify the 3D graphics to draw, with 3D projection
ax=fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(x,y,z,cstride=4,rstride=4)
```

# Exercise- Write a Python program to create a pie chart of the popularity of programming languages


```python
import matplotlib.pyplot as plt
#Data to plot
languages='Python','Java','C#','PHP','Javascript','C++'
popularity=[22.2,17.6,8.8,8,7.7,6.7]

#explode 1st slice
explode=(0.1,0,0,0,0,0)
plt.pie(popularity,explode=explode,labels=languages,
        autopct='%1.1f%%',shadow=True,startangle=140)
```
